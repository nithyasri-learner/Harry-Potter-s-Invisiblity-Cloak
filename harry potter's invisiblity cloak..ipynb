```python
import cv2
import numpy as np
import time

#caputure video using webcamera
cap = cv2.VideoCapture(0)

time.sleep(2)
#initialize the background
background =0
count = 0
#capure the background using 30 iteration so that it can get the good background
for i in range(60):
    ret,background = cap.read()
    if ret == False:
        continue
#To ensure the code is running till the webcam is capturing the video   
while(cap.isOpened()):
    # capture the image to process on it
    ret, img = cap.read()
    img = np.flip(img,axis=1)
    kernel = np.ones((5,5),np.uint8)
    if not ret:
        break
    count = count + 1
    
    #for better result convert BGR to HSV
    hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
    #(HSV- Hue Saturation Value)(color,saturation,Brightness)
    #shade of red from range of 0 to 10
    lower_red = np.array([100,40,40]) 
    upper_red = np.array([100,255,255])
    mask1 = cv2.inRange(hsv,lower_red,upper_red)#seperating the cloak part
    #shade of red from range 170 to180
    lower_red = np.array([155,40,40]) 
    upper_red = np.array([180,255,255])
    mask2 = cv2.inRange(hsv,lower_red,upper_red)
    #by adding we can get the red from both the values
    mask1=mask1+mask2
    #remove noise 
    mask1 = cv2.morphologyEx(mask1,cv2.MORPH_OPEN,kernel,iterations =2)
    
    mask1 = cv2.morphologyEx(mask1,cv2.MORPH_DILATE,kernel)
    
    mask2 =cv2.bitwise_not(mask1)#except the cloak
    
    res1 = cv2.bitwise_and(background,background,mask=mask1)#segmentation of the color
    res2 = cv2.bitwise_and(img,img,mask=mask2)#substitute the cloak part
    final_output = cv2.addWeighted(res1,1,res2,1,0)
    
    cv2.imshow('output',final_output)
    k = cv2.waitKey(0)
    if k==27:
        break
cap.release()
cv2.destroyAllWindows()
    
    
    
    
    
    
    
    
    
```
